```python
import numpy as np
np.random.seed(25)

token_ids = np.array([
    21, 39, 14, 62, 8,
    55, 47, 19, 83, 26,
    71, 34, 12, 58, 90,
    16, 44, 29, 67, 5
])

sequence_length = 4

X = []
Y = []

for i in range(len(token_ids) - sequence_length):
    X.append(token_ids[i:i + sequence_length])
    Y.append(token_ids[i + sequence_length])

X = np.array(X)
Y = np.array(Y)

print("Input Shape :", X.shape)
print("Target Shape:", Y.shape)

vocab_size = 100
embedding_dim = 12

embedding_matrix = np.random.randn(vocab_size, embedding_dim)

embedded = embedding_matrix[X]

seq_len = sequence_length

position = np.arange(seq_len)[:, np.newaxis]

div_term = np.exp(
    np.arange(0, embedding_dim, 2) *
    (-np.log(10000.0) / embedding_dim)
)

PE = np.zeros((seq_len, embedding_dim))
PE[:, 0::2] = np.sin(position * div_term)
PE[:, 1::2] = np.cos(position * div_term)

embedded = embedded + PE

mask = np.triu(np.ones((seq_len, seq_len)), k=1)
mask = mask * (-1e9)

Q = embedded
K = embedded
V = embedded

scores = np.matmul(Q, K.transpose(0, 2, 1))
scores = scores / np.sqrt(embedding_dim)
scores = scores + mask

scores = scores - np.max(scores, axis=-1, keepdims=True)

attention = np.exp(scores)
attention = attention / np.sum(attention, axis=-1, keepdims=True)

attention_output = np.matmul(attention, V)

W = np.random.randn(embedding_dim, vocab_size)

logits = np.matmul(attention_output, W)
logits = logits - np.max(logits, axis=-1, keepdims=True)

exp_logits = np.exp(logits)

probabilities = exp_logits / np.sum(exp_logits, axis=-1, keepdims=True)

last_probs = probabilities[:, -1, :]

loss = -np.mean(np.log(last_probs[np.arange(len(Y)), Y] + 1e-9))

predicted_tokens = np.argmax(last_probs, axis=1)

print("\nBackpropagation Step")
print("Weight update would be performed here during training.")

print("\n========== RESULTS ==========")
print("Loss:", loss)

print("\nFirst Input Sequence:")
print(X[0])

print("\nActual Next Token:")
print(Y[0])

print("\nPredicted Next Token:")
print(predicted_tokens[0])

print("\nAttention Output Shape:", attention_output.shape)
print("Probability Shape:", probabilities.shape)

print("\nARLM Forward Pass Completed Successfully!")
```
