```python
import numpy as np

# Read the corpus
file = open("corpus.txt", "r")
data = file.read().split()
file.close()

data = [word.lower() for word in data]

print("Corpus:", data)

# Create vocabulary
vocab = {}

for word in data:
    token = " ".join(list(word)) + " </w>"
    vocab[token] = vocab.get(token, 0) + 1

# Display vocabulary
print("\nVocabulary")
for word, count in vocab.items():
    print(word, ":", count)

# NumPy statistics
values = np.array(list(vocab.values()))
print("\nTotal Words:", np.sum(values))
print("Maximum Frequency:", np.max(values))


# Count adjacent pairs
def pair_frequency(words):

    count = {}

    for word, freq in words.items():

        letters = word.split()

        for i in range(len(letters)-1):

            p = (letters[i], letters[i+1])

            if p not in count:
                count[p] = freq
            else:
                count[p] += freq

    return count


# Replace the selected pair
def update_vocab(words, pair):

    result = {}

    old = pair[0] + " " + pair[1]
    new = pair[0] + pair[1]

    for word, freq in words.items():

        changed = word.replace(old, new)

        result[changed] = freq

    return result


# Learn merge rules
rules = []

for step in range(10):

    pairs = pair_frequency(vocab)

    if len(pairs) == 0:
        break

    highest = sorted(pairs.items(), key=lambda x: x[1], reverse=True)[0]

    print("\nMerge", step+1)
    print("Best Pair:", highest[0])
    print("Frequency:", highest[1])

    rules.append(highest[0])

    vocab = update_vocab(vocab, highest[0])


# Final vocabulary
tokens = []

for word in vocab:
    tokens.extend(word.split())

tokens = sorted(set(tokens))

print("\nFinal Vocabulary")
print(tokens)

print("Vocabulary Size:", len(tokens))


# Encoding
def encode(text):

    output = list(text)
    output.append("</w>")

    for rule in rules:

        i = 0

        while i < len(output)-1:

            if output[i] == rule[0] and output[i+1] == rule[1]:

                output[i] = output[i] + output[i+1]
                output.pop(i+1)

            else:
                i += 1

    return output


# Decoding
def decode(tokens):

    text = ""

    for token in tokens:
        text += token

    return text.replace("</w>", "")


# Testing
samples = ["newest", "lowest", "lower"]

print("\nTesting")

for word in samples:

    code = encode(word)

    text = decode(code)

    print("\nWord:", word)
    print("Encoded:", code)
    print("Decoded:", text)


print("\nMerge Rules")

for i in range(len(rules)):
    print(i+1, rules[i])
```
